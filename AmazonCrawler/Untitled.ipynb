```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
response = requests.get('https://www.amazon.jobs/en/locations/irvine-california?offset=0&result_limit=10&sort=relevant&category[]=software-development&distanceType=Mi&radius=24km&latitude=&longitude=&loc_group_id=&loc_query=&base_query=&city=&country=&region=&county=&query_options=&')
```

```python
response
```

```python
html_data = response.text
```

```python
data = BeautifulSoup(html_data, 'html.parser')
```

```python
print(data.prettify())
```

```python
jobs = data.find_all(class_ = "job-title")
```
