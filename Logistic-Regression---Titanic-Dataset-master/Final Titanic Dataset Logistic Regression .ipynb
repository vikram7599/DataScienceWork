```python
from sklearn.linear_model import LogisticRegression
import numpy as np 
import pandas as pd
```

```python
train_data=pd.read_csv("training_titanic_x_y_train.csv")
#train_data.reset_index(inplace=True)
train_data.head(10)
```

```python
train_data.isnull().sum()
```

```python
train_d = train_data.copy()
train_d["Age"].fillna(train_data["Age"].median(skipna=True), inplace=True)
train_d["Embarked"].fillna(train_data['Embarked'].value_counts().idxmax(), inplace=True)
train_d.drop('Cabin', axis=1, inplace=True)
train_d.head()
```

```python
train_d.isnull().sum()
```

```python
train_d['TravelAlone']=np.where((train_d["SibSp"]+train_d["Parch"])>0, 0, 1) #important---->cond,false true
train_d.drop('SibSp', axis=1, inplace=True)
train_d.drop('Parch', axis=1, inplace=True)
train_d.head()
```

```python
training=pd.get_dummies(train_d, columns=["Pclass","Embarked","Sex"])
training.drop('Sex_female', axis=1, inplace=True)
training.drop('Name', axis=1, inplace=True)
training.drop('Ticket', axis=1, inplace=True)

final_train = training
final_train.head()
```

```python
test_d=pd.read_csv("test_titanic_x_test.csv")
test_d.head()
```

```python
test_data = test_d.copy()
test_data["Age"].fillna(train_d["Age"].median(skipna=True), inplace=True)
test_data["Fare"].fillna(train_d["Fare"].median(skipna=True), inplace=True)
test_data.drop('Cabin', axis=1, inplace=True)

test_data['TravelAlone']=np.where((test_data["SibSp"]+test_data["Parch"])>0, 0, 1)

test_data.drop('SibSp', axis=1, inplace=True)
test_data.drop('Parch', axis=1, inplace=True)

testing = pd.get_dummies(test_data, columns=["Pclass","Embarked","Sex"])
testing.drop('Sex_female', axis=1, inplace=True)

testing.drop('Name', axis=1, inplace=True)
testing.drop('Ticket', axis=1, inplace=True)

final_test = testing
final_test.head()
```

```python
final_train['IsMinor']=np.where(final_train['Age']<=16, 1, 0)
final_train.head()
```
