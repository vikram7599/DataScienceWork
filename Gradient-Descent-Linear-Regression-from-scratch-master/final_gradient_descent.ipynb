```python
import numpy as np
import pandas as pd
tr_df = np.genfromtxt("boston_train.csv", delimiter = ",")
tr_df.shape[1]
```

```python
def step_grad(points, learn_rate, M):
    N = points.shape[0]
    num_col = points.shape[1]
    
    new_M = np.zeros(num_col)
    
    for i in range(N):
        x = points[i, 0:num_col-1]
        x = np.append(x, 1)
        y = points[i, num_col-1]
        for j in range(num_col):
            new_M[j] += (-2/N) * (y - (M * x).sum()) * x[j]
        M = M - (learn_rate * new_M)
    
    return M
```

```python
def gd(points, learn_rate, num_iter):
    num_col = points.shape[1]
    M = np.zeros(num_col)
    
    for i in range(num_iter):
        M = step_grad(points, learn_rate, M)

    print(i, "Cost= ", cost(points, M))
    return M
```

```python
def cost(points, M):
    total_cost = 0
    
    N = points.shape[0]
    num_col = points.shape[1]
    
    for i in range(N):
        x = points[i,0:num_col-1]
        x = np.append(x, 1)
        y = points[i,num_col-1]
        total_cost += (y - (M * x).sum()) ** 2
        
    total_cost = (1/N) * total_cost
    return total_cost
```

```python
def run():
    #= pd.read_csv("/Users/shagunsaboo/Downloads/trainb.csv")
    learning_rate = 0.00001
    num_iterations = 1000
    m = gd(tr_df, learning_rate, num_iterations)
    #print(m, c)
    return m[0:13], m[13]
```

```python
m,c=run()
print(m,c)
```

```python
test_data = np.genfromtxt("boston_test.csv", delimiter = ",")
```

```python
def predict(x, m, c):
    N = x.shape[0]
    y = np.zeros(N)
    
    for i in range(N):
        X = x[i,:]
        y[i] = (m * X).sum() + c
        
    return y
```

```python
x_test = test_data
```

```python
y_pred = predict(x_test, m, c)
```
