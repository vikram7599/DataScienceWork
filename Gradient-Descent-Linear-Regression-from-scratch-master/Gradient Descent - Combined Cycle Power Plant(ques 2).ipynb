## Generic Gradient Descent

```python
def step_grad(points,learn_rate,M):
    N=points.shape[0]
    num_col = points.shape[1]
    
    new_M=np.zeros(num_col)
    
    for i in range(N):
        x=points[i,0:num_col-1]
        x=np.append(x,1)
        y=points[i,num_col-1]
        for j in range(num_col):
            new_M[j] += (-2/N) * (y-(M *x).sum()) * x[j]
        M = M-(learn_rate * new_M)
        
    return M
```

```python
def grad_desc(points,learn_rate, num_iter):
    num_col = points.shape[1]
    M=np.zeros(num_col)
    for i in range(num_iter):
        M = step_grad(points, learn_rate, M)
        if i % 100 == 0:
            print(i, "cost=  ",cost(points,M))
    print(i,"cost=  ",cost(points, M))
    return M
```

```python
def cost(points,M):
    total_cost = 0
    
    N=points.shape[0]
    num_col = points.shape[1]
    
    for i in range(N):
        x=points[i,0:num_col-1]
        x=np.append(x,1)
        y=points[i,num_col-1]
        total_cost += (y - (M * x).sum()) ** 2
    
    total_cost = (1/N) * total_cost
    return total_cost
```

```python
def run():
    data=train_d[:,0:train_d.shape[1]]
    #change learn_rate and num_iter to get better results;0.001, 1000
    learn_rate=0.0001
    num_iter = 1000
    m = grad_desc(data,learn_rate, num_iter)
    return m[0:train_d.shape[1]-1],m[train_d.shape[1]-1]
```

```python
m,c=run()
print(m, c)
```

```python
test_data=np.genfromtxt("new_test_ccpp_x_test.csv",delimiter = ",")
```

```python
def predict(x,m,c):
    N = x.shape[0]
    y=np.zeros(N)
    
    for i in range(N):
        X = x[i,:]
        y[i] = (m * X).sum() + c
        
    return y
```
