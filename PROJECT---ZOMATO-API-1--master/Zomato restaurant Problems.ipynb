### Zomato Restaurant 2.3

Find out whether online table reservation is supported in 736 AD or not using Zomato API.

Print 'yes' if online table reservation is supported else 'no'

Note :

In Zomato response JSON. 1=Yes and 0 = No

```python
import requests

header = {"Accept":"application/json","user-key":"cf1194b048c4e62fcccfc0f02638d919", 'User-agent' : 'curl/7/43.0'}

res_id = 18241524
params = {'res_id' : res_id}
response = requests.get('https://developers.zomato.com/api/v2.1/restaurant', headers = header, params = params)
data = response.json()

if data.get('is_table_reservation_supported') :
    print('yes')
else:
    print('no')
```

### Zomato Restaurant 2.4

Fetch the latest review of restaurant 736 AD using Zomato API.

Print the name of user ,user rating and review text.

Note :

Due to limitation of basic API plan, it can only fetch 5 latest review

```python
import requests

header = {"Accept":"application/json","user-key":"cf1194b048c4e62fcccfc0f02638d919", 'User-agent' : 'curl/7/43.0'}

res_id = 18241524
params = {'res_id' : res_id}
response = requests.get('https://developers.zomato.com/api/v2.1/reviews', headers = header, params = params)
data = response.json()
for i in data['user_reviews'] :
    print(i.get('review').get('user').get('name'), ' ', i.get('review').get('rating'), ' ', i.get('review').get('review_text'))
```
