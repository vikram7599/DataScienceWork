```python
driver.get('https://etherscan.io/txs')
differentPageLink = []

for i in range(100):
    currPageLinks = []
    currPageElemes = driver.find_elements_by_class_name('myFnExpandBox_searchVal')
    for link in currPageElemes:
        currPageLinks.append(link.get_attribute('href'))
    differentPageLink.append(currPageLinks)
#     driver.execute_script('window.scrollBy(0, 2000);')
    nextButton = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_topPageDiv"]/nav/ul/li[4]/a')
    driver.get(nextButton.get_attribute('href'))
    time.sleep(1)
```

```python
import sys
```

```python
def getData(linksList):
    
    dataValues = []
    
    for link in linksList:
    
        currPageData = []
        driver.get(link)
        time.sleep(2)

            driver.execute_script('window.scrollBy(0, 1000);')
#         time.sleep(2)
        
        
        expandButton = driver.find_element_by_xpath('//*[@id="collapsedLink"]')
        expandButton.click()
        time.sleep(1)
        driver.execute_script('window.scrollBy(0, -1000);')
        time.sleep(1)
        
        
        try:
            txnHash = driver.find_element_by_xpath('//*[@id="spanTxHash"]').text
            currPageData.append(txnHash)
        except:
            currPageData.append("Remove")
        
        
        try:
            successData = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_maintable"]/div[2]/div[2]/span').text
            currPageData.append(successData)
        except:
            currPageData.append("Remove")
               

        try:
            timeData = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_maintable"]/div[4]/div[2]/span[2]').text
            currPageData.append(timeData[2:])
        except:
            currPageData.append("Remove")
        
        driver.execute_script('window.scrollBy(0, 600);')
        
        try:
            value = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_spanValue"]/span').text
            currPageData.append(value)
        except:
            currPageData.append("Remove")
        
        
        try:
            transactionFees = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_spanTxFee"]/span').text
            currPageData.append(transactionFees)
        except:
            currPageData.append("Remove")
        
        
        
        try:
            gasPrice = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_spanGasPrice"]').text
            currPageData.append(gasPrice)
        except:
            currPageData.append("Remove")
        
        
        
        try:
            gasLimit = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_spanGasLimit"]').text
            currPageData.append(gasLimit)
        except:
            currPageData.append("Remove")
        

        try:
            gasUsePercent = driver.find_element_by_xpath('//*[@id="ContentPlaceHolder1_spanGasUsedByTxn"]').text
            currPageData.append(gasUsePercent)
        except:
            currPageData.append("Remove")

        

        dataValues.append(currPageData)
        time.sleep(2)
    
    return dataValues
    
```

```python
for i in range(80, 100):
    currPageLinks = differentPageLink[i]
#     currPageData = []
    
    dataValue = getData(currPageLinks)
    
    columnName = ['Transaction Hash', 'Status', 'Time Consumed', 'Value', 'transactionFees', 'gasPrice', 'gasLimit', 'gasUsePercent']
    df = pd.DataFrame(dataValue, columns = columnName)
    df.to_csv('Page No.' + str(i+1) + '_1.csv')
    
```
