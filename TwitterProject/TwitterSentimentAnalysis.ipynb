```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
from nltk.corpus import wordnet
from nltk import pos_tag
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string

stop = stopwords.words('english')
punctuations = list(string.punctuation)
stop = stop + punctuations
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stop:
            pos1 = pos_tag([w]) #Pass pos tag as an array and not a single word because post tag expects an array
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos1[0][1])) #0th entry to get tuple and 1st entry for the pos
            output_words.append(clean_word.lower())
            #print(get_simple_pos(pos_tag([w])[0][1]))
    return output_words
```
