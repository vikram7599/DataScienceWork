# Store data in CSV

Fetch title , url of each book , no. of copies in stock and price of each book
So, visit each book webpage and make a list and then a dataframe to store in a csv file

```python
from bs4 import BeautifulSoup
import requests
```

```python
# currently writing code to extract details of first book only

response = requests.get('http://books.toscrape.com/')
data = BeautifulSoup(response.text, 'html.parser')
data
```

```python
# extracting url of first book

b1 = data.find(class_ = 'product_pod')
b1
```

```python
base_url = 'http://books.toscrape.com/'
b1_url = base_url + b1.h3.a['href']
print(b1_url)
```

```python
# loading the html of 1st book page

response = requests.get(b1_url)
data = BeautifulSoup(response.text, 'html.parser')
data
```

```python
# extracting title, price , no. of copies in stock

title = data.h1.string
price = data.find(class_ = 'price_color').string
qty = data.find(class_ = 'instock availability') # can't write .string as 2 children here 1 i tag ang 1 string so would return
# none here so, use .contents to get list from which last element and strip that to get ans.
qty = qty.contents[-1].strip()
```

```python
print(title)
print(b1_url)
print(price)
print(qty)
```

```python
# changing way to represent price and quantlty 
# we want only numbers to be shown
# have to do string processing on them to get desired data
```

#### Taking help of   regx  expressions 
#### so importing re library

```python
import re 
```

```python
# pass pattern for which we wanna check and the i/p string to search() function
# we wanna check if any number available in string extract that and ignore all the alphabets
# in reqular expressions there is a special alphabet - \d meanings numbers from 0-9
# so, to extract all numbers

re.search('\d', qty)
```

```python
# using .group fucntion of re library to get the matched outcome

re.search('\d', qty).group()
# it returns the first num that matches given pattern
```

```python
# now , \d - 0-9 nos
# and , \d+ means all the numbers in string that matches the pattern

qty = int(re.search('\d+', qty).group())
```

```python
# Now, working on extracting price numbers
# not including decimal here
re.search('\d+', price).group()
```

```python
# [] this rep all char we wanna match with either \d or . and + for all digits 
re.search('[\d.]', price).group()
```

```python
# this will give string as o/p string 
price = re.search('[\d.]+', price).group()
print(type(price))
```

```python
# converting to float 
# typecasting it

price = float(re.search('[\d.]+', price).group())
```

```python
# Now our data is extracted finally,

print(title)
print(b1_url)
print(price)
print(qty)
```

## EXTRACT DETAILS OF ALL 1000 BOOKS

```python
all_urls
```

```python
base_url = 'http://books.toscrape.com/catalogue/'
all_books_urls = []

for i in all_urls:
    res = requests.get(i)
    soup = BeautifulSoup(res.text, 'html.parser')
    book = soup.find_all(class_ = 'product_pod')
    
    for j in book:
        book_url = base_url + j.h3.a['href']
        all_books_urls.append(book_url)
```

```python
all_books_urls
```

```python
import re

title = []
prices = []
quantity = []

for book in all_books_urls:
    r = requests.get(book)
    soup = BeautifulSoup(r.text, 'html.parser')
    title.append(soup.h1.string)
    price = soup.find(class_ = 'price_color').string
    qty = soup.find(class_ = 'instock availability')
    qty = qty.contents[-1].strip()
    prices.append(float(re.search('[\d.]+', price).group()))
    quantity.append(int(re.search('\d+', qty).group()))
```

```python
all_book_details = []

for i in range(len(all_books_urls)):
    all_book_details.append([title[i], all_books_urls[i], prices[i], quantity[i]])
all_book_details
```

```python
import pandas as pd

df = pd.DataFrame(all_book_details, columns = ['Title', 'Link', 'Price', 'QuantityInStock'])
df
```

```python
df.to_csv('all-books.csv', index = False )
```
